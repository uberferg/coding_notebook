```python
%pylab inline
import pylab
import numpy as np
```

## Plot a function
### Using `matplotlib`

```python
x = numpy.linspace(-np.pi*3,np.pi*3,1000) #1000 linearly spaced numbers
f_x = numpy.sin(x) #computing the values of the function f(x) = sin(x)

#compose plot
pylab.plot(x,f_x)
pylab.show()
```

#### Adding additional functions

```python
g_x = numpy.cos(x) #compute g(x) = cos(x)
h_x = numpy.sin(x)/x #compute h(x) = sin(x)/x

pylab.plot(x, f_x) #re-plot f(x) bc it was cleared by the last show()
pylab.plot(x, g_x, x, h_x) #plot g(x), h(x)
pylab.show()
```
